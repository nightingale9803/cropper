```python
import cv2 as cv
import numpy as np
from math import *
from PIL import Image
from matplotlib import pyplot as plt

import os, sys
sys.path.append("..")

from modules.cropper import *
```

```python
img_path = 'test.png'
image = RubbingCropper(img_path)
# plot the two images (binary, gray) into one figure
plt.figure(figsize=(25, 15))
cnts = image.contours
print(f'Found {len(cnts)} contours in this image')
# draw contours onto binary for debugging
canvas = np.full_like(image.gray, 255)
cv.drawContours(canvas, cnts, -1, (0, 255, 0), 6)
plt.subplot(131), plt.imshow(image.binary, cmap='gray')
plt.title('Binary Image Inverted'), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(canvas, cmap='gray')
plt.title('Detected Contours'), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(image.gray, cmap='gray')
plt.title('Cleaned Image'), plt.xticks([]), plt.yticks([])
```
