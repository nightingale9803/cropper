```python
import os, sys
import matplotlib.pyplot as plt
from math import *

import pdfplumber

import os, sys

sys.path.append("..")
pdf_path = 'test_ocr_page.pdf'
```

```python
with pdfplumber.open(pdf_path) as pdf:
  # plot image
  image = pdf.pages[0].to_image()
  plt.figure(figsize=(15, 15))
  plt.imshow(image.original)
```

```python
with pdfplumber.open(pdf_path) as pdf:
    page = pdf.pages[0]
    # get width and height of the pdf page
    print(page.width, page.height)
    # extract words, can set a higher x_tolerance to merge adjcant numbers if there is problems
    words = page.extract_words(x_tolerance=8)
    # number text, there might be useless text like "典宾"、"宾一" which does not matter
    texts = [word["text"] for word in words]
    center_points = [
        (
            round((word["x0"] + word["x1"]) / 2),
            round((word["top"] + word["bottom"]) / 2),
        )
        for word in words
    ]
    print(texts)
    print(center_points)
```
